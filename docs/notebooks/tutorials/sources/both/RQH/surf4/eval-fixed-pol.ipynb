```python
for i in range(0,len(sNv)):

    p_M = sMv[0]
    p_N = sNv[i]

    
    soln = np.load('soln_M_' + str(p_M) + '_N_' + str(p_N) + '.npy')
    error = np.load('error_M_' + str(p_M) + '_N_' + str(p_N) + '.npy')
    bn = np.load('bn_error_M_' + str(p_M) + '_N_' + str(p_N) + '.npy')
        
    #G_chi.append(soln[-1])
    b_chi.append(error)
    #bn_chi.append(bn)
    max_I.append(max(abs(soln)))
```
