```python
N = 20 # Terms toa pproximate the infinite series

# Numer of dipoles
sMv = np.asarray([20,20,20,20])
sNv = np.asarray([10,20,30,40])

#sMv = np.asarray([10,])
#sNv = np.asarray([5,])

#np.asarray([25])
#sNv = sMv

b_chi = []
Bn_chi = []
max_I = []
K_chi = []
```

```python
from find_dips import K_decomp, alt_grid, bn_res
```

```python
for i in range(0,len(sMv)):

    p_M = sMv[i]
    p_N = sNv[i]
    
    #x = jnp.ones( ( p_M * 2 + 1 ) * ( p_N * 2 + 1 ) * 2 + 0 )

    theta = jnp.linspace(2 * jnp.pi * (1 / (p_M * 2 + 1)) * 1/2,
                         2 * jnp.pi * (1 - 1 / (p_M * 2 + 1) * 1/2),
                         p_M * 2 + 1)

    zeta = jnp.linspace(2 * jnp.pi / surf_winding.NFP * (1 / (p_N * 2 + 1)) * 1/2,
                        2 * jnp.pi / surf_winding.NFP * (1 - 1 / (p_N * 2 + 1) * 1/2),
                        p_N * 2 + 1)

    # Find grids for dipoles
    d_grid = alt_grid(theta,zeta) # dipole grid
    d_data = surf_winding.compute(["theta","zeta","e^theta_s","e^zeta_s",
                                   "x",'K',
                                   ], grid = d_grid) # dipole data

    
    d_data = iso_coords_interp(name, d_data, d_grid, surf_winding) # dipole data with isothermal coordinates
    
    d_data['K.e_u'] = dot(d_data['K'],d_data['e_u'])
    d_data['K.e_v'] = dot(d_data['K'],d_data['e_v'])
    
    soln = K_decomp(p_M, p_N,
                   d_data,
                   d_grid,
                   surf_winding,
                   dt, dz,
                   N,
                   d0)
    
    K_d = K_dips(p_M, p_N,
                 sdata, 
                 sdata2,
                 sdata3,
                 sgrid, surf_winding,
                 soln,
                 dt,dz, N, d0)

    B_d = _compute_magnetic_field_from_Current(sgrid, 
                                                K_d,
                                                surf_winding, eq, egrid, basis="rpz")
    B0 = B_d - B_s
    B_total = B_d + B_sec
    Bn_total = dot(edata["n_rho"],B_total)
    K0 = K_d - K_sv
    
    plot_figure2( Bn_total * dot(B_total,B_total)**(-1/2), egrid,
                 ''r' $ \frac{ (\mathbf{B_{sec} + B_{dip}}) \cdot \mathbf{n} }{| \mathbf{B_{sec} + B_{dip}} | }$ ' )
    
    plot_figure2( dot(B_d,B_d)**(1/2), egrid, ''r' $ | \mathbf{B_{dip}} | $ ' )
    
    plot_figure2( dot(B0,B0)**(1/2) * dot(B_s,B_s)**(-1/2), 
                 egrid, 
                 ''r' $ \frac{| \mathbf{B_s - B_{dip}}  |}{|\mathbf{B_s}|} $ ' )
    
    plot_figure2( dot(K_d,K_d)**(1/2), sgrid, ''r' $ | \mathbf{K} | $ ' )
    
    error = surf_int( dot(B0,B0) ** 2, edata, egrid )
    kerror = surf_int( dot(K0,K0) ** 2, sdata, sgrid )
    
    b_chi.append(error)
    K_chi.append(kerror)
    Bn_chi.append( max(abs(Bn_total * dot(B_total,B_total)**(-1/2))) )
    max_I.append(max(abs(soln)))
    
    np.save('soln_M_' + str(sMv[i]) + '_N_' + str(sNv[i]) + '.npy' ,soln)
    np.save('error_M_' + str(sMv[i]) + '_N_' + str(sNv[i]) + '.npy' , kerror)
    np.save('kerror_M_' + str(sMv[i]) + '_N_' + str(sNv[i]) + '.npy' , error)
    np.save('maxbn_M_' + str(sMv[i]) + '_N_' + str(sNv[i]) + '.npy' , max(abs(Bn_total * dot(B_total,B_total)**(-1/2))) )
```
